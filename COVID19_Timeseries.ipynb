```python
import json
import requests
import pandas as pd
import pprint
import numpy as np
```

```python
#REQUESTING COVID DATA--1
url_2 = 'https://data.covid19india.org/v4/min/timeseries.min.json' 
response_2=requests.get(url_2)
#READING THE DATA 
data_2=json.loads(response_2.text)
#pprint.pprint(data_1)
#pprint.pprint(data_1)
```

```python
df=pd.DataFrame(data_2)
df
```

```python
#transpose
df1=df.T
df1.head(2)
```

```python
df1=df1.dropna(subset='dates')
```

```python
df1['dates'].keys()
```

```python
l1=[]
l2=[]
for i in df1['dates']:
    for j in i:
        l1.append(j)
        l2.append(i[j])


```

```python
#Data_Merge
df2=pd.DataFrame([l1,l2])
df2=df2.T
df2
```

```python
df2.rename( columns={0:'dates'}, inplace=True)
df2_date=df1.set_index(df1['dates'].keys())
df2_date
```

```python
#df1.set_index(df1['dates'].keys())
```

```python
date=pd.DataFrame([x for x in df2[1]])
date=date.set_index(df2['dates'].keys())
date
```

```python
t1=[]
t2=[]
t3=[]
for i,y in df1['dates'].items():
    for j in y:
        t1.append(i)
        t2.append(j)
        t3.append(y[j])
```

```python
df=pd.DataFrame([t1,t2,t3])
df=df.T
df
```

```python
df.rename( columns={0:'statecode',1:'dates'},inplace=True)
df1_date = df.set_index(['statecode','dates'])
df1_date
```

```python
date=pd.DataFrame([x for x in df[2]])
date
```

```python
df1_date[2].keys()
```

```python
date=pd.DataFrame([i for i in df1_date[2]],index=df1_date[2].keys())
date
```
