```python
import requests
import json
import pandas as pd
import csv
```

```python
url = 'https://data.covid19india.org/v4/min/data.min.json'
response = requests.get(url)
```

```python
dict_1 = json.loads(response.text)        #to convert json to dict
print(type(dict_1))                       #to check the data type
```

```python
df = pd.DataFrame(dict_1)
df=df.transpose()
df
```

# DISTRICTS

```python
df1=df.dropna(subset='districts')
```

```python
l1 =[]
l2 =[]
l3 =[]
for i,y in df1['districts'].items():
    for j in y:
        l1.append(i)
        l2.append(j)
        l3.append(y[j])
df2 = pd.DataFrame([l1,l2,l3])
df2 = df2.T
df2.rename( columns={0:'statecode', 1 :'districts'}, inplace=True )
```

```python
df2['districts'].is_unique
df2[(df2['districts'].duplicated())]
df2[(df2['districts']=='Bilaspur')]
```

```python
df_districts =df2.set_index(['statecode','districts'])
df_districts 
```

```python
#method 2 instead of drop
t1 =[]
t2 =[]
for i in df1['districts']:
    for j in i:
        t1.append(j)
        t2.append(i[j])
dt2 = pd.DataFrame.from_dict([t1,t2])
dt2 = dt2.T
dt2.rename( columns={0 :'districts'}, inplace=True )
```

```python
districts = pd.DataFrame([x for x in df_districts[2]],index=df_districts[2].keys())
districts
```
