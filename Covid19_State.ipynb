```python
import json
import requests
import pandas as pd
import pprint
```

```python
#REQUESTING COVID DATA--1
url_1 = 'https://data.covid19india.org/v4/min/data.min.json' 
response_1=requests.get(url_1)
#READING THE DATA 
data_1=json.loads(response_1.text)
#pprint.pprint(data_1)
#pprint.pprint(data_1)
```

```python
df=pd.DataFrame(data_1)
df
```

```python
#transpose
df1=df.T
df1
```

```python
 pd.DataFrame([x for x in df1['delta']]).keys()
```

```python
df1['delta'].keys()
```

# delta

```python
delta = pd.DataFrame([x for x in df1['delta']],index=df1['delta'].keys())
```

```python
# DATA CLEANING - delta
```

```python
pd.isnull(delta).sum()
```

```python
#replacing nan with zero
delta=delta.fillna(0)
delta
```

```python
# checking dtype 
delta.dtypes
```

```python
# changing datatype of the all columns together
delta=delta.astype(int)
delta.dtypes
```

```python
delta.to_csv(r'D:\delta.csv')
```
